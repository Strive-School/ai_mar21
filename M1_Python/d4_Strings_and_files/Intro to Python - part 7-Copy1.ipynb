### Exercise 11.2

Do the same thing as you did for the previous exercise, but now process the text line by line. This is something that you would have to do if you had to process a very long text (as it is the case).

```python
# Counting words line by line.

# declares the file to check
file = open("text-files/blakepoems.txt")

def count_words_by_line(textdoc):
    # Reads each line of the inputted text document
    buffer = file.readlines()
    
    # Stores the counts for each word
    word_dict = {}
    
    # Loops thorugh each line in the buffer
    for line in buffer:
        
        #Ignores runtime of filtering text on blank lines
        if line != '\n':
            
            #Replaces all unwanted symbols with white space
            replace_symbols = line.replace("[", " ").replace("]", " ").replace(".", " ")\
            .replace(",", " ").replace(";", " ").replace(":", " ").replace("\"", " ")\
            .replace("!", " ").replace("-", " ")
            
            # Splits line into each word removing white space
            replace_space = replace_symbols.split(" ")
            
            # Removes any empty indexes generated by adding and splitting on space.
            cut_empty = [word for word in replace_space if word != ""]
            
            #loops through and removes line breaks that are parts of strings
            for word in cut_empty:
                
                # gives the final word to check in lowercase to avoid duplicates
                check_word = word.replace("\n", "").lower()
                
                # Updates if key exists or creates if it does not. 
                if check_word in word_dict.keys():
                    word_dict[check_word] += 1
                else:
                    word_dict[check_word] = 1
        
    return word_dict

#run the count.
#print(count_words_by_line(file))

## Easier print for accuracy check - uncomment as necessary
for word_count in count_words_by_line(file).items():
    print(word_count)
```
